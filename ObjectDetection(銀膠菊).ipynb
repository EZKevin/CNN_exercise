```python
imgFolder = "XMF"
if not os.path.exists(imgFolder):
    os.makedirs(imgFolder)
else:
    print("Folder opened")
    
files = os.listdir(imgFolder)
#print(files)

des = os.path.realpath(imgFolder).replace("\\","/")
print(des)
```

```python
dataDir = "C:/Users/user/Desktop/RawData/"
directory = os.listdir(dataDir)
#print(directory)

position = 0
for file in directory:
    print("Proccessing file: ",file)
    if ".jpg" in file:
        #look for labeled image
        if (file.replace(".jpg","")) == (directory[position+1].replace(".xml","")):
            #print(file," has a labeled file")
            #copy the labeled image only to destination folder 
            if shutil.copy(dataDir+file,des) and shutil.copy(dataDir+directory[position+1],des):
                print(file," and ",directory[position+1]," has copied to ",des+"\n")
        else: print(file,"was not labeled...Skipping....")
    else: print("Skipped non JPG file")
    position += 1
```
